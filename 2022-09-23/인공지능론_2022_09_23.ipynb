```python
import tensorflow as tf
import numpy
import matplotlib as mpl
import matplotlib.pyplot as plt

X = [1., 2., 3.]

# tf.Variable: 그래프를 계산하면서 최적화할 변수
# tf.random_normal: 각 변수들의 초기 값을 정규분포 랜덥값으로 초기화
w = tf.Variable(tf.random.normal([1], dtype=tf.float32))
b = tf.Variable(tf.random.normal([1], dtype=tf.float32))

@tf.function
def forward(x1):
  return x1 * w + b # 1차식

Y = forward(X)

print(Y)
```

```python
plt.plot(X, Y, 'ro')
plt.plot(X,Y)
plt.show()
```

```python
import tensorflow as tf
import matplotlib as mpl
import matplotlib.pyplot as plt

x = tf.linspace(-3.0, 3, 1000)
y1 = tf.nn.sigmoid(x) # 시그모이드 함수 값
y2 = tf.nn.softplus(x) # 소프트플러스 함수 값
y3 = tf.nn.softsign(x) # 소프트 사인 함수 값

plt.plot(x, y1, "-", label="sigmoid")
plt.plot(x, y2, "--", label="softplus")
plt.plot(x, y3, ":", label="softsign")
plt.xlabel("x")
plt.title("sigmoid, softplus, softsign")
plt.legend()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x_data = np.arange(-2.0, 2.0, 0.1)
y_data = []
W = tf.Variable(0.3)
b = tf.Variable(0.1)

y_data = W * x_data + b


plt.plot(x_data, y_data, "ro", label="list degree")
plt.ylim([0.0,0.7])
plt.xlim([-2.0,2.0])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_data,y_data)
plt.legend()
plt.show()
```

```python
y_data = W * x_data * x_data + b
plt.plot(x_data, y_data, "ro", label="list degree")
plt.ylim([0.0,0.7])
plt.xlim([-2.0,2.0])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_data,y_data)
plt.legend()
plt.show()
```

```python
y_data = W * x_data * x_data * x_data + b
plt.plot(x_data, y_data, "ro", label="list degree")
plt.ylim([0.0,0.7])
plt.xlim([-2.0,2.0])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_data,y_data)
plt.legend()
plt.show()
```
